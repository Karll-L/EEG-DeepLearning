```python
#File 1
for y in range(0, 16):
    print("-----Subject " + str(y) + "----\n")
    signals, signal_headers, header = highlevel.read_edf('Subject' + str(y) + '_1.edf')
    print("File Subject" + str(y) + "_2.edf Read Successful\n")
    
    #print("Signal Frequency is "+ str(signal_headers[0]['sample_rate'])) 
    sfreq = signal_headers[0]['sample_rate']
    for x in range(0, 21):
        FT = np.fft.fft(signals[x])/len(signals[x])
        FT = FT[range(int(len(signals[x])/2))]
        print(FT)
        print("\n")
        
        mag = []
        
        #save array into a csv file
        for z in range(0, 15500):
            mag.append(math.sqrt(FT.real[z]**2 + FT.imag[z]**2))
        string = "FFT"+str(y)+"_Signalfull"+str(x)+"_1.csv"
        np.savetxt(string, mag, delimiter=' ')
        print(string + " created successfully\n")
        #np.savetxt('FFT-Subject' + str(y) + '_2-Signal'+ str(x+1) + '.csv', mag, delimiter=' ')
        #print("File FFT-Subject" + str(y) + "_2-Signal"+ str(x+1) + ".csv created successfully\n")

    print("\n")
    print("\n")
```
