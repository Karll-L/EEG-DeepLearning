```python
from pyedflib import highlevel
#this is another method to process edf files
```

```python
#File 1
for y in range(0, 16):
    print("-----Subject " + str(y) + "----\n")
    signals, signal_headers, header = highlevel.read_edf('Subject' + str(y) + '_1.edf')
    print("File Subject" + str(y) + "_1.edf Read Successful\n")
    #print("Signal Frequency is "+ str(signal_headers[0]['sample_rate'])) 
    sfreq = signal_headers[0]['sample_rate']
    for x in range(0, 21):
        FT = np.fft.fft(signals[x])/len(signals[x])
        FT = FT[range(int(len(signals[x])/2))]
        print(FT)
        print("\n")
        #save array into a csv file
        np.savetxt('FFT-Subject' + str(y) + '_1-Signal'+ str(x+1) + '.csv', FT, delimiter=' ')
        print("File FFT-Subject" + str(y) + "_1-Signal"+ str(x+1) + ".csv created successfully\n")

    print("\n")
    print("\n")
```

```python
#File 2
for y in range(0, 16):
    print("-----Subject " + str(y) + "----\n")
    signals, signal_headers, header = highlevel.read_edf('Subject' + str(y) + '_2.edf')
    print("File Subject" + str(y) + "_2.edf Read Successful\n")
    #print("Signal Frequency is "+ str(signal_headers[0]['sample_rate'])) 
    sfreq = signal_headers[0]['sample_rate']
    for x in range(0, 21):
        FT = np.fft.fft(signals[x])/len(signals[x])
        FT = FT[range(int(len(signals[x])/2))]
        print(FT)
        print("\n")
        #save array into a csv file
        np.savetxt('FFT-Subject' + str(y) + '_2-Signal'+ str(x+1) + '.csv', FT, delimiter=' ')
        print("File FFT-Subject" + str(y) + "_2-Signal"+ str(x+1) + ".csv created successfully\n")

    print("\n")
    print("\n")
```

```python
for x in range(0, 16):

    signals, signal_headers, header = highlevel.read_edf('Subject'+str(x)+'_1.edf')
    #print("Signal Frequency is "+ str(signal_headers[0]['sample_rate'])) 
    sfreq = signal_headers[0]['sample_rate']

    fourierTransform = np.fft.fft(signals[0])/len(signals[0])           # Normalize amplitude
    #print("Before: " + str(fourierTransform))

    fourierTransform = fourierTransform[range(int(len(signals[0])/2))] # Exclude sampling frequency
    print("FFT of file Subject"+str(x)+"_1.edf" ": " + str(fourierTransform))

    #save array into a csv file
    np.savetxt('FFTofSubject'+str(x)+'_1.csv', fourierTransform, delimiter=",")
```
